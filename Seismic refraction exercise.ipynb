```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
with open('PROFILE1_forward.txt') as f:    #import data for forward profile
    array_d1 = [[x for x in line.split()] for line in f]
down_1 = np.array(array_d1)

x = []
y = []

for line in down_1:    #store profile position and time
    x.append(int(line[0]))
    y.append(float(line[1]))

plt.scatter(x,y,label='forward')    #plot forward profile data

x_1 = x[0:5]    #store forward profile data into two section, 1 is direct wave, 2 is refracted head wave
x_2 = x[4:]

y_1 = y[0:5]
y_2 = y[4:]

m,c = np.polyfit(x_1, y_1, 1)    #calculate the slope m for direct wave
plt.plot(x_1, m*np.array(x_1)+c, color='black')    #plot direct wave data

n,b = np.polyfit(x_2, y_2, 1)    #calculate the slope n for refracted head wave
plt.plot(x_2, n*np.array(x_2)+b, color='black')    #plot refracted head wave data

y_int_x = [x[0],x[4]]
y_int = [n*x[0]+b,y[4]]
plt.plot(y_int_x, y_int, 'k--')    #extend refracted head wave line to y-int, denoting td
plt.annotate("td", (y_int_x[0], y_int[0]))

td = y_int[0]
v1d = 1/m*10**3    #calculate v1d by 1/slope of direct wave
v2d = 1/n*10**3    #calculate v2d by 1/slope of refracted head wave
print("td =", "%.2f" % td, "ms")
print("v1d =", "%.2f" % v1d, "m/s")
print("v2d =", "%.2f" % v2d, "m/s")

with open('PROFILE1_reverse.txt') as g:    #import data for reverse profile
    array_u1 = [[x for x in line.split()] for line in g]
up_1 = np.array(array_u1)

x = []
y = []

for line in up_1:    #store profile position and time
    x.append(int(line[0]))
    y.append(float(line[1]))

plt.scatter(x,y,label='reverse')    #plot reverse profile data

x_1 = x[0:4]    #store reverse profile data into two section, 1 is direct wave, 2 is refracted head wave
x_2 = x[3:]

y_1 = y[0:4]
y_2 = y[3:]

m,c = np.polyfit(x_1, y_1, 1)    #calculate the slope m for direct wave
plt.plot(x_1, m*np.array(x_1)+c, color='black')    #plot direct wave data

n,b = np.polyfit(x_2, y_2, 1)    #calculate the slope n for refracted head wave
plt.plot(x_2, n*np.array(x_2)+b, color='black')    #plot refracted head wave data
 
y_int_x = [x[0],x[3]]
y_int = [n*x[0]+b,y[3]]
plt.plot(y_int_x, y_int, 'k--')    #extend refracted head wave line to y-int, denoting tu
plt.annotate("tu", (y_int_x[0], y_int[0]))

tu = y_int[0]
v1u = np.abs(1/m*10**3)    #calculate v1u by 1/slope of direct wave
v2u = np.abs(1/n*10**3)    #calculate v2u by 1/slope of refracted head wave
print("tu =", "%.2f" % tu, "ms")
print("v1u =", "%.2f" % v1u, "m/s")
print("v2u =", "%.2f" % v2u, "m/s")

v1 = (v1d+v1u)/2    #calculate the v1

ic = 0.5*(math.asin(v1/v2d)+math.asin(v1/v2u))    #calculate the critical angle
print("critical angle =", "%.2f" % (ic*180/math.pi), "degree")

dip = 0.5*(math.asin(v1/v2d)-math.asin(v1/v2u))    #calculate the dip
print("dip =", "%.2f" % (dip*180/math.pi), "degree")

v2 = v1/math.sin(ic)   #calculate the v2
print("v1 =", "%.2f" % v1, "m/s")
print("v2 =", "%.2f" % v2, "m/s")

Hd = td*v1/2/np.cos(ic)    
Hu = tu*v1/2/np.cos(ic)
print("Hd =", "%.2f" % Hd, "m")
print("Hu =", "%.2f" % Hu, "m")

plt.yscale('linear')
plt.xscale('linear')
plt.title('Forward and reverse profile 1')
plt.xlim([x[-1],x[0]])
plt.xlabel('profile position')
plt.ylabel('time (ms)')
plt.legend()
plt.show()

```
